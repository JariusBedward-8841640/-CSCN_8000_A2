# Imports

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Lasso, Ridge, LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

# Question 1: Data Preprocessing (Missing Values, Scaling, Outliers)

```python
print (" Question 1: Data Preprocessing")

columns = ["CRIM", "ZN", "INDUS" 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

df =  pd.read_csv("./data/housing.csv", sep=r"\s+", header=None, names=columns)
print(f"Dataset loaded successfully. Shape: {df.shape}")

# Check for the missing values

missing_value_count = df.isnull().sum().sum()
print(f"Total Missing Values: {missing_value_count}")

# Check for outliers using IQR Method
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
outliers_per_col = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
print("\nOutlier Counts per column (IQR Method):")
print(outliers_per_col[outliers_per_col > 0])

#Feature Scaling (Standardization)
x = df.drop(columns= ["MEDV"])
y = df["MEDV"]

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
print("\n Feature scaling complete")
```

# Question 2: Use 80% of samples as traning data size

```python
print(" Question 2: Train-Test Split")
x_train, x_test, y_train, y_test = train_test_split(
    x_scaled, y, test_size=0.2, random_state=42
)
print(f"Training Data Size: {x_train.shape[0]} samples (80%)")
print(f"Test Data Size: {x_test.shape[0]} samples (20%)")
```

# Question 3, 4 & 6: Model Training, Predicintg, and Evaluation

```python
print("\n Question 3, 4 & 6: Model Training & Evaluation")

#Define the models

models = {
    "Linear Regression": LinearRegression(),
    "XGBoost": XGBRegressor(random_state=42, n_estimators=100, learning_rate=0.1),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42, n_estimators=100),
    "L1 Regularization (Lasso)": Lasso(alpha=0.1, random_state=42),
    "L2 Regularization (Ridge)": Lasso(alpha=1.0, random_state=42)

}

predictions = {}
metrics_list = []

# Evaluate stanrdard regression models
for name, model in models.items():
    # Train model
    model.fit(x_train, y_train)

    # Predict on test data
    preds = model.predict(x_test)
    predictions[name] = preds

    # Meassure accuracies and error metrics
    mae = mean_absolute_error(y_test, preds)
    rmse = np.sqrt(mean_absolute_error(y_test, preds))
    r2 = r2_score(y_test, preds)

    metrics_list.append({"Model": name, "R2": r2, "MAE": mae, "RMSE": rmse})
    print(f"[{name}] -> R-Squared: {r2:.4f} | MAE: ${mae:.4f} | RMSE: ${rmse:.4f}")

    #special handling for logistic regression because it is a classification algorithim so we have to do this to prevent sklearn from crashing on float targets
print("\n Running Logistic Regression")
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(x_train, y_train.astype(int))
log_preds = log_reg.predict(x_test)
predictions["Logistic Regression"] = log_preds

mae_lr = mean_absolute_error(y_test, log_preds)
rmse_lr = np.sqrt(mean_squared_error(y_test, log_preds))
r2_lr = r2_score(y_test, log_preds)
metrics_list.append({"Model": "Logistic Regression","R2": r2_lr, "MAE": mae_lr, "RMSE": rmse_lr})
print(f"[Logistic Regression] -> R-Squared: {r2_lr:.4f} | MAE: ${mae_lr:.4f} | RMSE: ${rmse_lr:.4f}")

```

# Question 6 (Continued): Finding the best Model

```python
metrics_df = pd.DataFrame(metrics_list)
best_model_info =  metrics_df.loc[metrics_df["R2"].idmax()]
best_model_name =  best_model_info["Model"]
best_model_obj = models[best_model_name] if best_model_name in models else log_reg

print(f"\nBest Model Name: {best_model_name}")
print(f"Highest Accuracy (R-Squared): {best_model_info["R2"]:.4f} ({best_model_info["R2"]*100:.2f}%)")
print(f"Mean Absolute Error (MAE):   ${best_model_info["MAE"]:.4f}($1,000s)")
print(f"Root Mean Squared Error: ${best_model_info["RMSE"]:.4f}($1,000s)")
```
